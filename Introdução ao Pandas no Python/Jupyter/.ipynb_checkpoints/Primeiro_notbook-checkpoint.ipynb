## Resumos de Visualização de Dados simples e úteis
- `.head ()`
    - Retorna as primeiras linhas do dataframe. Por padrão, mostra as primeiras 5 linhas, mas você pode especificar o número de linhas que deseja ver , como `df.head (10)` para as primeiras 10 linhas.
- `.shape`
    - Fornece a dimensão do dataframe na forma de uma tupla `(n_linhas, n_colunas)`. É útil para entender rapidamente o tamanho do seu dataframe.
- `.describe ()`
    - Gera estatísticas descritivas que resumem a tendêncial central, dispersão e forma da distribuição de um conjunto de dados. Inclui contagem, média, desvio padrão, valores mínimos e máximos, e percitis (25%, 50%, 75%) para colunas numérias.

```python
display (vendas_df.head (10))
display (vendas_df.shape)
display (vendas_df.describe ())
```

## Edição dos dados
### Pergar apenas dados de uma coluna
- Esse trabalho de pegar dados de uma coluna nada mais é trabalhar com Series do panda, cada caluna retirada de um dataframe é uma serie ou seja `pd.Series`

```python
produtos = vendas_df ["Produto"] # Dentro de [] colocamos o nome da coluna que desejamos pegar os dados
display (produtos)
```

```python
produtos = vendas_df [["Produto", "ID Loja"]]
display (produtos)
```

## .loc, um método muito importante
- Pegar 1 linha
- Pegar linhas de acordo com alguma condição
- Pegar linhas e olunas específicas
- Pegar 1 valor específico

### Pegar uma ou varias linhas

```python
display (vendas_df.loc[1]) # Devera pegar os dados da primeira linha
display (vendas_df.loc[0:5]) # Devera pegar da linha 0 até a 5
```

### Pegar linhas que correspondem a uma condição

```python
display (vendas_df.loc[vendas_df["ID Loja"] == "Norte Shopping"]) # Pega apenas as linhas em que o ID da loja corresponde ao Norte Shopping
```

### Pegar várias linhas e colunas usando o loc
Basicamente a sintax do loc é: `dataframe_df.loc[linhas, colunas]`

```python
vendas_norteshopping_df = vendas_df.loc[vendas_df["ID Loja"] == "Norte Shopping", ["ID Loja", "Produto", "Quantidade"]]
# Este código devera pegar as linhas em que o ID Loja é igual a Norte Shopping, mas mostrar apenas as colunas ID Loja, Produto e Quantidade
display (vendas_norteshopping_df)
```
